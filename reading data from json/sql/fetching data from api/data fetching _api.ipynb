```python
import pandas as pd
import requests
```

```python
url = "https://streaming-availability.p.rapidapi.com/shows/top"

querystring = {
    "country": "us",
    "service": "netflix",
    "show_type": "movie"
}

headers = {
    "X-RapidAPI-Key": "a0c69fac50mshcee6c191e044c13p18a80ejsn38c7cdfd52f9",
    "X-RapidAPI-Host": "streaming-availability.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

if response.status_code == 200:
    print(response.json())
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
pd.DataFrame(response.json())
```

```python
df =pd.DataFrame()
```

```python
df
```

```python
for i in range(1,11):
    querystring = {
        "country": "us",
        "service": "netflix",
        "show_type": "movie",
        "page": i
    }
    response = requests.get(url, headers=headers, params=querystring)
    if response.status_code == 200:
        df = pd.concat([df, pd.DataFrame(response.json())], ignore_index=True)
    else:
        print(f"Error: {response.status_code} - {response.text}")
```
