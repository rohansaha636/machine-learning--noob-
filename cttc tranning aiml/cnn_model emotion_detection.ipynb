```python
import keras
import pickle
import tensorflow
import numpy as np
```

```python
#loading the data from pickle files:

with open(r"C:\Documents and Settings\rohan\machine learning (noob)\cttc tranning aiml\images.p",'rb') as f:
    X=pickle.load(f)

with open(r"C:\Documents and Settings\rohan\machine learning (noob)\cttc tranning aiml\label.p",'rb') as f:
    y=pickle.load(f)
```

```python
print("Images:" ,X.shape)
print("Lables:" ,y.shape)
```

```python
#label encoding output
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()

y=le.fit_transform(y)
y
```

```python
#standarizing the images for cnn model

X=X/255
X=X.reshape(-1,48,48,1)
```

```python
#using one - hot encoding  to output col:
from keras.utils import to_categorical
y=to_categorical(y)
```

```python
#spliting tranning and testing datas:
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(X,y,test_size=0.2)
```

```python
print("Xtrain:" ,xtrain.shape)
print("ytrain:" ,ytrain.shape)
print("Xtest:" ,xtest.shape)
print("ytest:" ,ytest.shape)
```

```python
#cnn model training:
from keras.models import Sequential
from keras.optimizers import Adam
from keras.layers import Dense , Conv2D, MaxPooling2D,Flatten
```

```python
model =Sequential()
# 1st convolutional layer:
model.add(Conv2D(45,(3,3), activation ='relu',input_shape=(48,48,1)))

model.add(MaxPooling2D(pool_size=(2,2)))

#2nd Convoluitional layer:
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#3rd covolutional layer:
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

#adding faltten layer: converts 2d to 1d
model.add(Flatten())

#hidden layer
model.add(Dense(256,activation='relu'))

#output layer:
model.add(Dense(5, activation='softmax'))
model.compile(Adam(learning_rate=0.001),
              loss= 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist =model.fit(xtrain,ytrain,epochs =20, validation_data=(xtest,ytest))
```

```python
# to save the model for deyployment:
# .h5 -> HDF5 FILE FORMAT
# it store model's optimizer's state model architecture, traning configurations, weights

model.save('emotion_detection_model.h5')
print("Model saved")
```
