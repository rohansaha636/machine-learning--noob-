```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import os # help to interact with os 
import pickle #help to zip/compress the img file in byte (binary)
```

```python
#acessing the image present inside "emotion"
#getcwd:- fetching the current working directory 
img_dir =os.path.join(os.getcwd(),"Emotion")
```

```python
print(os.listdir(os.getcwd()))
```

```python
# to checking the directory where it is currently , we are working
print(os.getcwd())
```

```python
# to store all the the input (image) and the output (label) in diff []
images =[]
labels =[]
```

```python
#performing image preprocessing and saving the files 
#listdir : to list all folders/file present in any directory
for i in os.listdir(img_dir): #to iterate over 'Emotion' folder
    for j in os.listdir(os.path.join(img_dir,i)):# to ilterate over the sub folder [angry,sad] etc
        img_path= os.path.join(os.path.join(img_dir,i),j) #to fetch the img path
        img =cv2.imread(img_path) #to read the img
        img =cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #TO  convert 3 colour channel (bcr) to 1 channel(grayscale)
        img =cv2.resize(img,(48,48)) #as all img have the shaper of 48 x48
        images.append(img) #to store processed images in the "images" list
        labels.append(i.split(' ')[0]) # to store the files  name (ex sad, angry etc )

    print(i,'--->completed ')
```

```python
set(labels)
```

```python
images =np.array(images)
labels =np.array(labels)
```

```python
#pickle the images and labels files 
# 'wb' -  write biniary  as pickle file store only byte data
#'images.p' -->'images'-> files anmes ,'.p' -> pickle file,s extenstion
img_final = os.getcwd()
with open(os.path.join(img_final,'images.p'), 'wb')as f:
    pickle.dump(images,f)

with open(os.path.join(img_final,'label.p'),'wb')as f:
    pickle.dump(labels,f)
```
