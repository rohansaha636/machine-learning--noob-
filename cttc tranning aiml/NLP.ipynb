### vectorization is the process of conveting the words into nmbers
1. count vectorizator 
2. tf-idf vectorizator

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
txt=[ "i love dogs and i love cats"]
print(txt)
```

```python
x=cv.fit_transform(txt)
x
```

```python
print (x)
```

```python
cv.vocabulary_
```

```python
print(x.toarray())
```

```python
txt=[ "i love painting and i love ice cream"]
```

```python
y=cv.fit_transform(txt)
print(y)
```

```python
print(y)
```

```python
cv.vocabulary_
```

```python
print(y.toarray())
```

```python
##tf-idf stands for (term frequency and inverse document frequency) it is used to find the importance of a word in a document
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
text=["the ant killed the elephant"]
z=tf.fit_transform(text)
z
```

```python
tf.vocabulary_
```

```python
print(z.toarray())
```
