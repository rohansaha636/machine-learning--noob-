CAPPING

```python

upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
lower_limit
```

```python

df['cgpa'] = np.where(
    df['cgpa']>upper_limit,
    upper_limit,
    np.where(
        df['cgpa']<lower_limit,
        lower_limit,
        df['cgpa']
    )
)
```

np.where() is a powerful NumPy function that performs conditional operations on arrays, offering two main modes: finding indices that meet conditions or creating new arrays with conditional value replacement.
